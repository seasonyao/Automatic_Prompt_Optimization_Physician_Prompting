```python
# Group rows by 'Group'
grouped_df = valA.groupby('section_header')

# Create separate DataFrames for each group
separate_dfs = {}
for group_name, group_df in grouped_df:
    separate_dfs[group_name] = group_df.copy()

# Display the separate DataFrames
print("\nSeparate DataFrames:")
for group_name, group_df in separate_dfs.items():
    print(f"\nsection_header: {group_name}")
    print(group_df)
```

```python
# Path to the zip file
# zip_file_path = 'gpt_preds.zip'
zip_file_path = 'best_prompts_preds.zip'

# Extract the contents of the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall('extracted_folder')

# Initialize a list to store the arrays for each folder
folder_arrays = []

# Get the list of folder names in the extracted_folder
folder_names = [folder for folder in os.listdir('extracted_folder') if os.path.isdir(os.path.join('extracted_folder', folder))]

# Iterate over each folder
for folder_name in folder_names:
    folder_path = os.path.join('extracted_folder', folder_name)

    # Get the list of .npy files in the folder and sort them alphabetically
    npy_files = sorted([file for file in os.listdir(folder_path) if file.endswith('.npy')])

    # Iterate over each .npy file in the folder and load the arrays
    folder_array = [np.load(os.path.join(folder_path, npy_file)) for npy_file in npy_files]

    # Append the folder_array to folder_arrays
    folder_arrays.append(folder_array)
```
