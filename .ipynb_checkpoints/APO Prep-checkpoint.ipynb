```python
import pandas as pd
```

```python
# df = pd.read_csv('gpt_best_prompts_summs.csv')
df = pd.read_csv('TaskA-TrainingSet.csv')
```

```python
# Initialize an empty DataFrame to store the final sampled rows
sampled_df = pd.DataFrame()

# Group by 'section_header' and sample rows from each group
for name, group in df.groupby('section_header'):
    n_rows = min(len(group), 10)  # Number of rows to sample, maximum 10
    sample = group.sample(n=n_rows, random_state=1)  # You can change the random_state for different randomness
    sampled_df = pd.concat([sampled_df, sample])

# Keep only the required columns
sampled_df = sampled_df[['ID','section_header', 'dialogue', 'section_text', 'gpt-4']]
```
